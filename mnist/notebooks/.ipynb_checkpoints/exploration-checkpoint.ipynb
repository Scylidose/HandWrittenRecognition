```python
import os

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn import manifold

import os

%matplotlib inline
```

```python
data = pd.read_csv("../input/mnist_784.csv")
```

```python
pixel_values = data.iloc[:, :-1]
targets = data.iloc[:, -1:]
targets = targets.astype(int)
```

```python
print(pixel_values.shape)
print(targets.shape)
```

```python
single_image = pixel_values.iloc[1].values.reshape(28, 28)
plt.imshow(single_image, cmap='gray')
```

```python
targets.value_counts()
```

```python
plt.hist(targets)
plt.xlabel("Handwritten number")
plt.ylabel("Count")
plt.show()
```

```python
tsne = manifold.TSNE(n_components=2, random_state=42)

transformed_data = tsne.fit_transform(pixel_values.iloc[:3000].values)
```

```python
tsne_df = pd.DataFrame(np.column_stack((transformed_data, targets[:3000])), columns=["x", "y", "targets"])

tsne_df.loc[:, "targets"] = tsne_df.targets.astype(int)
```

```python
tsne_df.head()
```

```python
grid = sns.FacetGrid(tsne_df, hue="targets", height=6, aspect=1.5)
grid.map(plt.scatter, "x", "y").add_legend()
```
